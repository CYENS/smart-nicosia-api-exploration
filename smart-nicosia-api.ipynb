```python
from smart_nicosia_api import get_devices, iter_unique_by_device_type, fetch
import json
import pandas as pd

data = get_devices(
  base_url="https://nokia.smartnicosia.eu/backend/openapi/getTenantDevices",
  entity_type="TENANT",
  entity_name="CYTA",
  insecure=True,  # only if you decide to skip TLS verification
)
unique_devices_examples  = list(iter_unique_by_device_type(data))
print(json.dumps(unique_devices_examples, indent=2))
```

```python
unique_devices_examples_df = pd.DataFrame(unique_devices_examples)
unique_devices_examples_df
```

```python
unique_device_types = pd.unique(unique_devices_examples_df.device_type)
unique_device_types
```
